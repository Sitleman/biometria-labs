```python
import cv2
import matplotlib.pyplot as plt
```

```python
path = r'images/persons.png'
image = cv2.imread(path, cv2.IMREAD_COLOR)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()
gray_scale_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(gray_scale_image, cmap='gray')
plt.show()
# print(gray_scale_image)
```

```python
haarcascade = cv2.CascadeClassifier('cascades/haarcascade_frontalface_default.xml')
faces = haarcascade.detectMultiScale(image, 1.1, 9)
faces
```

```python
for face in faces:
    cv2.rectangle(image, (face[0], face[1]), (face[0] + face[2], face[1] + face[3]), (255, 255, 0), 10)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()
```
